# Working With Dataset

Load Train and Test Datasets and set coloumn names

```python
import pandas as pd

column_names = [
    "duration", "protocol_type", "service", "flag", "src_bytes", "dst_bytes",
    "land", "wrong_fragment", "urgent", "hot", "num_failed_logins", "logged_in",
    "num_compromised", "root_shell", "su_attempted", "num_root", "num_file_creations",
    "num_shells", "num_access_files", "num_outbound_cmds", "is_host_login",
    "is_guest_login", "count", "srv_count", "serror_rate", "srv_serror_rate",
    "rerror_rate", "srv_rerror_rate", "same_srv_rate", "diff_srv_rate",
    "srv_diff_host_rate", "dst_host_count", "dst_host_srv_count",
    "dst_host_same_srv_rate", "dst_host_diff_srv_rate", "dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate", "dst_host_serror_rate", "dst_host_srv_serror_rate",
    "dst_host_rerror_rate", "dst_host_srv_rerror_rate", "class", "difficulty"
]

train_file_path = '/content/NetworkAnomalyDetectionDatasets/KDDTrain+.csv'
test_file_path = '/content/NetworkAnomalyDetectionDatasets/KDDTest+.csv'

# Read the CSV file into a pandas DataFrame
train_df = pd.read_csv(train_file_path, header=None, names=column_names)
test_df = pd.read_csv(test_file_path, header=None, names=column_names)

# Display the first 5 rows of the DataFrame
display(train_df.head())
display(test_df.head())
```

Check Missing Values In Train and Test Dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import Counter

print("Check Missing Values In Train Dataset")
print(train_df.isnull().sum().sum())

print("Check Missing Values In Test Dataset")
print(test_df.isnull().sum().sum())
```

Delete difficulty coloumn from datasets

```python
train_df = train_df.drop('difficulty', axis=1)
test_df = test_df.drop('difficulty', axis=1)

print("Delete difficulty coloumn from datasets")
display(train_df.head())
display(test_df.head())
```

Identifying Numerical and Categorical Features in Datasets

```python
# جدا کردن ستون 'class' (برچسب) از ویژگی ها
X_train = train_df.drop('class', axis=1)
y_train = train_df['class']

X_test = test_df.drop('class', axis=1)
y_test = test_df['class']

# شناسایی ویژگی‌های عددی و دسته‌ای
numerical_train_features = X_train.select_dtypes(include=[np.number]).columns
categorical_train_features = X_train.select_dtypes(include=[object]).columns

print(f"\nNumerical Features: {list(numerical_train_features)}")
print(f"Categorical Features: {list(categorical_train_features)}")
```

Encoding Categorical Features

```python
import pandas as pd
import numpy as np

# اعمال One-Hot Encoding برای ویژگی‌های دسته‌ای
# از get_dummies پانداس استفاده می کنیم که برای این کار مناسبه
X_train_encoded = pd.get_dummies(X_train, columns=categorical_train_features, drop_first=False)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_train_features, drop_first=False)

# اطمینان از اینکه تعداد ستون‌های X_train_encoded و X_test_encoded یکسان باشند.
# این مورد زمانی رخ میده که یک دسته خاص فقط در دیتاست آموزش یا تست وجود داشته باشه.
# باید ستون‌های گمشده در یکی از دیتافریم‌ها رو با 0 پر کنیم.
missing_cols_train = set(X_test_encoded.columns) - set(X_train_encoded.columns)
for c in missing_cols_train:
    X_train_encoded[c] = 0

missing_cols_test = set(X_train_encoded.columns) - set(X_test_encoded.columns)
for c in missing_cols_test:
    X_test_encoded[c] = 0

# مرتب کردن ستون ها برای اطمینان از اینکه هر دو DataFrame دارای ترتیب ستون های یکسان هستند
X_test_encoded = X_test_encoded[X_train_encoded.columns]


print("\n--- بعد از One-Hot Encoding ---")
display(X_train_encoded.head())
print(f"تعداد ستون‌ها بعد از Encoding: {X_train_encoded.shape[1]}")
```

Scaling Numerical Features

```python
# شناسایی ویژگی‌های عددی پس از One-Hot Encoding (که شامل ستون‌های جدید باینری نیستند)
# این کار رو روی X_train_encoded انجام می دهیم تا ستون های One-Hot شده رو در نظر نگیره
numerical_features_after_encoding = [col for col in numerical_train_features if col in X_train_encoded.columns] # اطمینان از حضور ستون ها

# تعریف Scaler
scaler = MinMaxScaler()

# اعمال Scaler روی داده‌های آموزش
X_train_scaled = X_train_encoded.copy() # ایجاد کپی برای جلوگیری از تغییر در DataFrame اصلی
X_train_scaled[numerical_features_after_encoding] = scaler.fit_transform(X_train_encoded[numerical_features_after_encoding])

# اعمال Scaler روی داده‌های تست (فقط Transform، نه Fit)
X_test_scaled = X_test_encoded.copy()
X_test_scaled[numerical_features_after_encoding] = scaler.transform(X_test_encoded[numerical_features_after_encoding])

print("\n--- بعد از نرمال‌سازی (Min-Max Scaling) ---")
display(X_train_scaled.head())
```

Label Encoding for 'class'

```python
# تبدیل انواع حملات به یک برچسب واحد 'attack'
# لیستی از انواع حملات در NSL-KDD (ممکنه نیاز به اصلاح جزئی داشته باشد اگر دیتاست شما شامل همه اینها نیست)
attack_types = ['apache2', 'back', 'buffer_overflow', 'ftp_write', 'guess_passwd', 'httptunnel',
                'imap', 'ipsweep', 'land', 'loadmodule', 'mstealth', 'multihop', 'named',
                'neptune', 'nmap', 'normal', 'perl', 'phf', 'pod', 'portsweep', 'rootkit',
                'satan', 'smurf', 'spy', 'sqlattack', 'teardrop', ' processtable', 'udpstorm',
                'warezclient', 'warezmaster', 'xlock', 'xsnoop', 'snmpgetattack', 'worm',
                'xterm', 'sendmail', 'saint', 'mailbomb', 'snmpguess', 'ps', 'syslog', 'ppmtelnet'] # این لیست کامل نیست و فقط برای ایده است.

# بهترین راه برای شناسایی برچسب های حمله:
# ابتدا ببینید چه مقادیری در ستون 'class' وجود دارند:
unique_classes_train = y_train.unique()
unique_classes_test = y_test.unique()
print(f"\nبرچسب‌های منحصر به فرد در دیتاست آموزش: {unique_classes_train}")
print(f"برچسب‌های منحصر به فرد در دیتاست تست: {unique_classes_test}")

# حالا برچسب 'normal' رو جدا می کنیم و بقیه رو به 'attack' تغییر می دهیم
y_train_binary = y_train.apply(lambda x: 'normal' if x == 'normal' else 'attack')
y_test_binary = y_test.apply(lambda x: 'normal' if x == 'normal' else 'attack')

# اعمال Label Encoding برای تبدیل 'normal' و 'attack' به 0 و 1
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_binary)
y_test_encoded = label_encoder.transform(y_test_binary) # از fit_transform برای تست استفاده نکنید

print("\n--- بعد از آماده‌سازی برچسب‌ها ---")
print(f"برچسب‌های آموزش (اولین ۵): {y_train_encoded[:5]}")
print(f"برچسب‌های تست (اولین ۵): {y_test_encoded[:5]}")
print(f"Mapping برچسب‌ها: {list(label_encoder.classes_)}") # مثلا: ['attack', 'normal'] -> [0, 1]
```

Handling Class Imbalance

```python
print("\n--- بررسی عدم تعادل کلاس قبل از مدیریت ---")
print("آموزش:", Counter(y_train_encoded))
print("تست:", Counter(y_test_encoded))
```

```python
X_train_final = X_train_scaled.values
X_test_final = X_test_scaled.values
y_train_final = y_train_encoded
y_test_final = y_test_encoded

print(f"\nشکل نهایی داده‌های آموزش (X_train_final): {X_train_final.shape}")
print(f"شکل نهایی برچسب‌های آموزش (y_train_final): {y_train_final.shape}")
print(f"شکل نهایی داده‌های تست (X_test_final): {X_test_final.shape}")
print(f"شکل نهایی برچسب‌های تست (y_test_final): {y_test_final.shape}")
```

# Neural Network Model Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.utils import class_weight # برای مدیریت عدم تعادل کلاس
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns

print(f"\nشکل نهایی داده‌های آموزش (X_train_final): {X_train_final.shape}")
print(f"شکل نهایی برچسب‌های آموزش (y_train_final): {y_train_final.shape}")
print(f"شکل نهایی داده‌های تست (X_test_final): {X_test_final.shape}")
print(f"شکل نهایی برچسب‌های تست (y_test_final): {y_test_final.shape}")
```

```python
# تعداد ویژگی‌های ورودی (تعداد ستون‌ها پس از پیش‌پردازش)
input_shape = X_train_final.shape[1]

# تعریف مدل Sequential
model = Sequential([
    # لایه ورودی و اولین لایه مخفی
    # تعداد نورون‌ها (units) معمولاً توانی از 2 در نظر گرفته میشه (64, 128, 256, 512, ...)
    # تابع فعال‌سازی 'relu' رایج‌ترین انتخاب برای لایه‌های مخفی است.
    Dense(units=128, activation='relu', input_shape=(input_shape,)),
    BatchNormalization(), # اضافه کردن Batch Normalization
    Dropout(0.2),         # اضافه کردن Dropout با نرخ 20%

    # لایه مخفی دوم
    Dense(units=64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    # لایه خروجی
    # برای طبقه‌بندی باینری (Normal/Attack)، 1 نورون با sigmoid
    Dense(units=1, activation='sigmoid')
])

# خلاصه مدل (Model Summary)
model.summary()
```
